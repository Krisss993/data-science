To nie jest odpowiedni df do grupowania

```python
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Product': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C'],
    'Sales': [150, 200, 130, 170, 300, 250, 220, 210, 400, 350, 380, 360],
    'Year': [2021, 2021, 2021, 2021, 2022, 2022, 2022, 2022, 2023, 2023, 2023, 2023]
}
df = pd.DataFrame(data)
```

```python
df.pivot_table( columns=['Region','Product'], aggfunc='mean', values='Sales').dropna()
```

```python
d = df.groupby(by='Region')
d.groups
```

```python
d.get_group('East')
```

Grupuje po regionach wyznaczając średnią z sales

```python
df.groupby(by='Region')['Sales'].mean()
```

```python
df.groupby('Year')['Sales'].mean()
```
